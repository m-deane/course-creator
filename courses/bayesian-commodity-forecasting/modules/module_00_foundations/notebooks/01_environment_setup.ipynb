## 4. PyMC - Our Primary Bayesian Framework

PyMC is the main probabilistic programming library we'll use throughout the course.

```python
import pymc as pm
import arviz as az

print(f"PyMC: {pm.__version__}")
print(f"ArviZ: {az.__version__}")

# Test basic model
print("\nRunning quick PyMC test (this may take a minute)...")

with pm.Model() as test_model:
    # Simple model: Normal with unknown mean
    mu = pm.Normal('mu', mu=0, sigma=1)
    obs = pm.Normal('obs', mu=mu, sigma=1, observed=np.random.randn(10))
    
    # Sample (just 100 draws for speed)
    trace = pm.sample(100, tune=100, cores=1, progressbar=False, random_seed=42)

print(f"Posterior mean: {trace.posterior['mu'].mean().values:.3f}")
print("✅ PyMC installation OK")
```

## 5. ArviZ - Bayesian Visualization and Diagnostics

ArviZ provides plotting and diagnostic tools for Bayesian inference.

```python
# Test ArviZ plotting
az.style.use('arviz-darkgrid')

# Plot trace from our test model
az.plot_trace(trace, var_names=['mu'])
plt.tight_layout()
plt.show()

# Summary statistics
summary = az.summary(trace, var_names=['mu'])
print(summary)

print("\n✅ ArviZ OK")
```

## 9. Final Verification

Run this cell to perform a comprehensive check.

```python
def run_verification():
    """Run all verification checks and report results."""
    checks = []
    
    # Python version
    major, minor = sys.version_info[:2]
    checks.append(("Python 3.11+", major == 3 and minor >= 11))
    
    # Core packages
    try:
        import numpy, pandas, scipy
        checks.append(("Core scientific stack", True))
    except ImportError:
        checks.append(("Core scientific stack", False))
    
    # Visualization
    try:
        import matplotlib, seaborn
        checks.append(("Visualization libraries", True))
    except ImportError:
        checks.append(("Visualization libraries", False))
    
    # Bayesian stack
    try:
        import pymc, arviz
        checks.append(("PyMC + ArviZ", True))
    except ImportError:
        checks.append(("PyMC + ArviZ", False))
    
    # Data retrieval
    try:
        import yfinance
        checks.append(("Data retrieval (yfinance)", True))
    except ImportError:
        checks.append(("Data retrieval (yfinance)", False))
    
    # Print results
    print("\n" + "=" * 50)
    print("VERIFICATION RESULTS")
    print("=" * 50)
    
    all_passed = True
    for name, passed in checks:
        status = "✅" if passed else "❌"
        print(f"{status} {name}")
        if not passed:
            all_passed = False
    
    print("=" * 50)
    
    if all_passed:
        print("\n🎉 All checks passed! You're ready for the course.")
    else:
        print("\n⚠️ Some checks failed. Please resolve issues above.")
    
    return all_passed

verification_passed = run_verification()
```
