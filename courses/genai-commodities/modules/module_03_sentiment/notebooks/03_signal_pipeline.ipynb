## 1. Signal Generation Framework

**Trading Signal Components:**

1. **Direction**: Long (bullish) or Short (bearish)
2. **Strength**: How strong is the signal (0-100)
3. **Confidence**: How confident are we (0-1)
4. **Entry Timing**: When to enter
5. **Exit Conditions**: Stop loss and take profit

**Signal Rules:**
- **Strong Buy**: Sentiment > +1.0, Consensus > 0.75
- **Buy**: Sentiment > +0.5, Consensus > 0.60
- **Neutral**: -0.5 < Sentiment < +0.5
- **Sell**: Sentiment < -0.5, Consensus > 0.60
- **Strong Sell**: Sentiment < -1.0, Consensus > 0.75
