## What Is Power Automate?

Power Automate is Microsoft's cloud-based workflow automation service that lets you connect applications and services through automated sequences called **flows** — without writing traditional server-side code. It is part of the **Microsoft Power Platform**, which also includes Power Apps, Power BI, Copilot Studio, and Power Pages, all sharing a common data layer (Dataverse) and connector library of over 1,000 pre-built integrations.

When something happens in one system — an email arrives, a form is submitted, a schedule fires — Power Automate can react, process data, and take action in another system automatically.

---

## Core Concepts

Everything in Power Automate is built from four primitives:

| Concept | One-line description |
|---------|---------------------|
| **Flow** | The unit of automation — a named sequence with one trigger and one or more actions |
| **Trigger** | The event that starts a flow (schedule, new email, HTTP request, button press) |
| **Action** | A single operation the flow performs (send email, create file, call API, post to Teams) |
| **Connector** | A pre-built adapter to an external service — wraps that service's API into named triggers and actions |

Every flow follows the same skeleton:

```
Trigger  →  [optional Logic]  →  Action(s)
```

Connectors supply both the triggers and the actions. Microsoft ships 1,000+ connectors (SharePoint, Outlook, Teams, SQL Server, Salesforce, HTTP, and hundreds more).

````python
import requests
import json

# Power Automate management API base URL
# All authenticated management operations (list flows, run flows, get run history)
# are scoped under this base URL.
POWER_AUTOMATE_API_BASE = "https://api.flow.microsoft.com"

# The Graph API base URL — used to authenticate and interact with Microsoft 365 data
# that flows act on (users, mail, SharePoint, Teams, etc.)
GRAPH_API_BASE = "https://graph.microsoft.com/v1.0"

print("Power Automate Management API base URL:")
print(f"  {POWER_AUTOMATE_API_BASE}")
print()
print("Microsoft Graph API base URL (M365 data layer):")
print(f"  {GRAPH_API_BASE}")
print()

# The metadata endpoint is publicly reachable and confirms the service is up
metadata_url = "https://api.flow.microsoft.com/providers/Microsoft.ProcessSimple"

response = requests.get(
    metadata_url,
    timeout=10,
    # Accept JSON — the API defaults to OData XML without this header
    headers={"Accept": "application/json"}
)

print(f"Discovery endpoint: {metadata_url}")
print(f"Status: {response.status_code} {response.reason}")

if response.status_code == 200:
    print()
    print("Connectivity confirmed. Power Automate management API is reachable.")
elif response.status_code in (401, 403):
    # Some endpoints require auth — a 401 still confirms the service is reachable
    print()
    print(f"Service reachable (auth required for this endpoint — expected).")
else:
    print(f"Unexpected status. Check your network connection.")
````

## Flow Types at a Glance

Power Automate has five flow types. The right type for a job depends on what starts the automation:

| Flow Type | What Starts It | Best For |
|-----------|---------------|----------|
| **Automated Cloud Flow** | An event in a connected service | Event-driven integrations, notifications |
| **Instant Cloud Flow** | A user pressing a button | On-demand tasks, user-initiated processes |
| **Scheduled Cloud Flow** | A time-based recurrence | Batch processing, daily reports |
| **Desktop Flow (RPA)** | A cloud flow or manual start | Legacy apps with no API, UI automation |
| **Business Process Flow** | A user moving through stages | Guided, compliance-driven human workflows |

The three cloud flow types (Automated, Instant, Scheduled) run entirely in Microsoft's cloud. Desktop flows require the **Power Automate Desktop** agent installed on a local machine.
