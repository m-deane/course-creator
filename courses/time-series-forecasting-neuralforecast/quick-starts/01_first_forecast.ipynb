## Load a Real Dataset

We use the **AirPassengers** dataset — monthly totals of international airline passengers (1949-1960). This is built into neuralforecast so no download is needed.

```python
from neuralforecast import NeuralForecast
from neuralforecast.models import NHITS
from neuralforecast.losses.pytorch import MQLoss
from neuralforecast.utils import AirPassengersDF

# The nixtla format: unique_id, ds, y
df = AirPassengersDF
print(f"Shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
print(f"Date range: {df['ds'].min()} to {df['ds'].max()}")
df.head()
```

## Train NHITS and Forecast

NHITS (Neural Hierarchical Interpolation for Time Series) is a strong default model. We train it with a probabilistic loss (MQLoss) to get prediction intervals.

```python
# Configure and train
horizon = 12  # Forecast 12 months ahead
models = [NHITS(h=horizon, input_size=24, max_steps=200,
                loss=MQLoss(level=[80, 90]),
                scaler_type="robust")]

nf = NeuralForecast(models=models, freq="MS")
nf.fit(df=df)

# Predict
forecasts = nf.predict()
forecasts.head()
```

## Visualize the Forecast

The output includes the point forecast (NHITS) and prediction intervals (lo/hi-80, lo/hi-90).

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, ax = plt.subplots(figsize=(12, 5))

# Plot historical data (last 36 months)
history = df.tail(36)
ax.plot(history["ds"], history["y"], "k-", label="Historical", linewidth=2)

# Plot forecast with prediction intervals
ax.plot(forecasts["ds"], forecasts["NHITS-median"], "b-", label="Forecast (median)", linewidth=2)
ax.fill_between(forecasts["ds"], forecasts["NHITS-lo-90"], forecasts["NHITS-hi-90"],
                alpha=0.15, color="blue", label="90% PI")
ax.fill_between(forecasts["ds"], forecasts["NHITS-lo-80"], forecasts["NHITS-hi-80"],
                alpha=0.3, color="blue", label="80% PI")

ax.set_title("NHITS Forecast with Prediction Intervals", fontsize=14)
ax.set_xlabel("Date")
ax.set_ylabel("Passengers")
ax.legend()
plt.tight_layout()
plt.show()
```
